```python
import pickle
import glob
import sys, copy
from astropy.table import Table
import numpy as np
import pyccl as ccl
import glob
import matplotlib.pyplot as plt
def load(filename, **kwargs):
    with open(filename, 'rb') as fin:
        return pickle.load(fin, **kwargs)
import itertools
```

```python
plt.figure(figsize=(12,4),dpi=100)
plt.subplot(121)
color = ['r','b','g']


file = ['flagship_summary_stat_DES_MoR_no_Mwl_scatter_Gamma1_gaussian_lensing_variable_Mwl.npy',
       'flagship_summary_stat_DES_MoR_Mwl_scatter_model_Gamma1_gaussian_lensing_variable_Mwl.npy',
       'flagship_summary_stat_DES_MoR_Mwl_scatter_model_Gamma0.7_gaussian_lensing_variable_Mwl.npy']
for i, f in enumerate(file): 
    print(f)
    data = np.load(f, allow_pickle=True).item()
    print(data.keys())
    plt.plot(data['mean_log10m200b'][:,0], ':', color=color[i],label = f.split('MoR_')[1], lw=4)

plt.legend(fontsize=8)
plt.xlabel('rich index')
plt.ylabel('mean log10M')
plt.xlim(-0.1, 4.1)
plt.grid(True)
plt.ylim(14, 15.9)

plt.subplot(122)
color = ['r','b','g']



file = ['flagship_like_summary_stat_DES_MoR_no_Mwl_scatter_Gamma1_gaussian_lensing_variable_Mwl.npy',
       'flagship_like_summary_stat_DES_MoR_Mwl_scatter_model_Gamma1_gaussian_lensing_variable_Mwl.npy',
       'flagship_like_summary_stat_DES_MoR_Mwl_scatter_model_Gamma0.7_gaussian_lensing_variable_Mwl.npy']
for i, f in enumerate(file): 
    print(f)
    data = np.load(f, allow_pickle=True).item()
    print(data.keys())
    plt.plot(data['mean_log10m200b'][:,0], ':', color=color[i],label = f.split('MoR_')[1], lw=4)

plt.legend(fontsize=8)
plt.xlabel('rich index')
plt.ylabel('mean log10M')
plt.xlim(-0.1, 4.1)
plt.grid(True)
plt.ylim(14, 15.9)
plt.savefig(f'sims_mass.png', dpi = 300, bbox_inches='tight',)
```
