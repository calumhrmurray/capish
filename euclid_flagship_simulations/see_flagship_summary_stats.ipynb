```python
import pickle
import glob
import sys, copy
from astropy.table import Table
import numpy as np
import pyccl as ccl
import glob
import matplotlib.pyplot as plt
def load(filename, **kwargs):
    with open(filename, 'rb') as fin:
        return pickle.load(fin, **kwargs)
import itertools
```

```python
plt.figure(dpi=100)
color = ['r','b','g']
file = ['flagship_summary_stat_DES_MoR_no_Mwl_scatter_Gamma1_gaussian_lensing_variable_log10Mwl.npy',
       'flagship_summary_stat_DES_MoR_Mwl_scatter_Gamma1_gaussian_lensing_variable_log10Mwl.npy',
       'flagship_summary_stat_DES_MoR_Mwl_scatter_Gamma0.7_gaussian_lensing_variable_log10Mwl.npy']
for i, f in enumerate(file): 
    print(f)
    data = np.load(f, allow_pickle=True).item()
    print(data.keys())
    plt.plot(data['mean_log10m200b'][:,0], color=color[i],label = f.split('MoR_')[1], lw=1)


file = ['flagship_summary_stat_DES_MoR_no_Mwl_scatter_Gamma1_gaussian_lensing_variable_Mwl.npy',
       'flagship_summary_stat_DES_MoR_Mwl_scatter_Gamma1_gaussian_lensing_variable_Mwl.npy',
       'flagship_summary_stat_DES_MoR_Mwl_scatter_Gamma0.7_gaussian_lensing_variable_Mwl.npy']
for i, f in enumerate(file): 
    print(f)
    data = np.load(f, allow_pickle=True).item()
    print(data.keys())
    plt.plot(data['mean_log10m200b'][:,0], ':', color=color[i],label = f.split('MoR_')[1], lw=4)

plt.legend(fontsize=8)
plt.xlabel('rich index')
plt.ylabel('mean log10M')
```
