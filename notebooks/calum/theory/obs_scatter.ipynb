```python
# equation 5
def P( lambda_obs , lambda_true , z , f_prj , delta_mu , sigma , tau ):
    mu = lambda_true + delta_mu
    a = ( 1 - f_prj ) * np.exp( -( lambda_obs - mu )**2 / ( 2 * sigma**2 ) )
    b = np.exp( 0.5 * tau * ( 2* mu + tau*sigma**2 - 2 * lambda_obs ) ) * torch.erfc( torch.tensor( ( mu + tau * sigma**2 - lambda_obs ) / ( np.sqrt(2) * sigma ) ) )
    return a + b 
```
