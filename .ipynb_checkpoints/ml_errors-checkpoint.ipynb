We can predict the weak lensing mass errors using a matched filter and assuming a galaxy density. This can give us an error for the obtained mass given the true signal, which can then be scattered....

Our matched filter is,

\begin{equation}
A = \frac{\tau_i C^{-1}_{ij} d_j}{\tau_i C^{-1}_{ij} \tau_j} = r_s
\end{equation}

where $\tau$ is the template, $C$ is the noise covariance and $d$ is the observation.

The variance on the amplitude can be calculated from the covariance,

\begin{equation}
\sigma^2 = \left( \tau C^{-1} \tau \right)^{-1}
\end{equation}

```python
import numpy as np
from scipy.integrate import quad

# Define the galaxy redshift distribution
def n_z(z, z0=0.5, beta=1):
    """
    Redshift distribution function n(z) ~ z^2 exp(-(z/z0)^beta).
    
    Parameters:
    - z : redshift
    - z0 : characteristic redshift
    - beta : shape parameter
    
    Returns:
    - n(z): unnormalized redshift distribution at redshift z
    """
    return z**2 * np.exp(-(z / z0)**beta)

# Function to compute the fraction of galaxies above a certain redshift z_min
def fraction_galaxies_above_z(z_min, z0=0.5, beta=1):
    """
    Computes the fraction of galaxies with redshift > z_min.
    
    Parameters:
    - z_min : minimum redshift
    - z0 : characteristic redshift of the distribution
    - beta : shape parameter for the exponential tail
    
    Returns:
    - fraction : fraction of galaxies above z_min
    """
    # Integrate the distribution from z_min to infinity
    numerator, _ = quad(n_z, z_min, np.inf, args=(z0, beta))
    
    # Integrate the full distribution from 0 to infinity to normalize
    denominator, _ = quad(n_z, 0, np.inf, args=(z0, beta))
    
    # The fraction is the ratio of the two integrals
    return numerator / denominator

# Example usage:
z_min = 1  # Minimum redshift for the fraction
fraction = fraction_galaxies_above_z(z_min, z0=0.5, beta=1)
print(f"Fraction of galaxies above z = {z_min}: {fraction:.4f}")
```

```python
import numpy as np
from scipy.special import gammaincc

def f(z_min, z0=0.5):
    """
    Computes the fraction of galaxies with redshift > z_min analytically using the incomplete gamma function.
    
    Parameters:
    - z_min : minimum redshift
    - z0 : characteristic redshift of the distribution
    
    Returns:
    - fraction : fraction of galaxies above z_min
    """
    # Use the normalized upper incomplete gamma function
    return gammaincc(3, z_min / z0)

# Example usage:
z_min = 0.7  # Minimum redshift for the fraction
fraction = f(z_min, z0=0.5)
print(f"Fraction of galaxies above z = {z_min}: {fraction:.4f}")
```

```python
def sigma_lnMwl( M_lens , z_lens , c_lens , n_gal ):
    

    R_bins = np.linspace(0.4, 4, 8)
    R = (R_bins[1:] + R_bins[:-1]) / 2.
    lens = tl.triaxial_lens(M_lens, c_lens, z_lens, 0, H0=67, Om0=0.319)

    dl = lens.cosmo.angular_diameter_distance(lens.z_lens).value
    theta = R / dl

    # ESD signal for M_lens
    esd_t = lens.esd_t(theta, 0)
    
    # calculate the number density of galaxies per arcminute squared, beyond this redshift...
    n_gal = n_gal * f( z_lens )

    # Number of galaxies in each radial bin
    Areas = np.pi * ((R_bins[1:] / dl)**2 - (R_bins[:-1] / dl)**2) * (180 / np.pi * 60)**2
    N_gal = Areas * n_gal

    # Covariance matrix
    sigma_cr = lens.sigma_cr( z_lens, z_lens + 0.5 )  # Critical surface density
    C = np.eye(len(R)) * (0.3**2 * sigma_cr**2 / N_gal)

    # Step 2: Calculate the derivative of the signal with respect to ln M_lens
    delta_lnM = 0.01  # Small perturbation
    M_lens_perturb = M_lens * (1 + delta_lnM)
    lens_perturb = tl.triaxial_lens(M_lens_perturb, c_lens, z_lens, 0, H0=67, Om0=0.319)

    # ESD signal for perturbed mass
    esd_t_perturb = lens_perturb.esd_t(theta, 0)

    # Numerical derivative of esd_t with respect to ln M_lens
    d_esd_d_lnM = (esd_t_perturb - esd_t) / delta_lnM

    # Step 3: Compute the Fisher information matrix
    F_lnM = d_esd_d_lnM.T @ np.linalg.inv(C) @ d_esd_d_lnM

    # Step 4: Calculate the uncertainty in ln M_lens
    sigma_lnM_wl = 1 / np.sqrt(F_lnM)

    # Output the uncertainty
    return sigma_lnM_wl
```

```python
sigma_lnM_wl_val = sigma_lnMwl( 1e14 , 0.5 , 5 , 30 )
sigma_lnM_wl_val
```

```python
# Step 5: Convert to uncertainty in log_10 M_lens
sigma_log10M_wl = sigma_lnM_wl_val / np.log(10)

# Output the uncertainty in log_10 M_lens
sigma_log10M_wl
```

```python
class ESD_simulator:
    
    def __init__(self ):
        """
        Initialize the Universe_simulation class.
        """
        self.R_bins = np.linspace(0.4, 4, 8)
        self.R = ( self.R_bins[1:] + self.R_bins[:-1] ) / 2.
        self.c_lens = 4
        self.n_gal = 30
        
    def esd_profile( self , param_values , lens , dl  ):
        """
        Calculate the noiseless esd profile
        """
        
        log10M_lens, z_lens = param_values
        z_lens = z_lens.numpy()
        
        theta = self.R / dl
        
        # ESD signal for M_lens
        esd_t = lens.esd_t( theta, 0 )
        
        return esd_t
        

    def run_simulation(self, param_values):
        """
        Run the simulation using the variable parameters provided.
        """
        
        log10M_lens, z_lens = param_values
        z_lens = z_lens.numpy()
        
        lens = tl.triaxial_lens( 10**log10M_lens, self.c_lens, z_lens, 0, H0=67, Om0=0.319)

        dl = lens.cosmo.angular_diameter_distance( lens.z_lens ).value
        
        # ESD signal for M_lens
        esd_t = self.esd_profile( param_values , lens , dl  )

        # calculate the number density of galaxies per arcminute squared, beyond this redshift...
        n_gal = self.n_gal * f( z_lens )
        
        # Number of galaxies in each radial bin
        Areas = np.pi * ((self.R_bins[1:] / dl)**2 - (self.R_bins[:-1] / dl)**2) * (180 / np.pi * 60)**2
        N_gal = Areas * n_gal

        # Covariance matrix
        sigma_cr = lens.sigma_cr( z_lens, z_lens + 0.5 )  # Critical surface density
        #C = np.eye(len(R)) * (0.3**2 * sigma_cr**2 / N_gal)
        
        # add errors to the esd profile
        err_realisation_esd_t = np.random.normal( loc = 0 , scale = 0.3 * sigma_cr / N_gal**0.5 )
        
        # perform a chi2, or alternatively just return the shear profile
        
        return np.concatenate([esd_t + err_realisation_esd_t, [z_lens]])

   
```

```python
# calculate an MCMC for comparison

def log_likelihood( param , esd_t_obs , z_lens , c_lens = 4 , R = ESD_full_profile.R , R_bins = ESD_full_profile.R_bins , n_gal = 30 ):
    
    log10M_lens = param

    lens = tl.triaxial_lens( 10**log10M_lens, c_lens, z_lens, 0, H0=67, Om0=0.319)

    dl = lens.cosmo.angular_diameter_distance(lens.z_lens).value
    theta = R / dl

    # ESD signal for M_lens
    esd_t = lens.esd_t( theta, 0 )

    # calculate the number density of galaxies per arcminute squared, beyond this redshift...
    n_gal = n_gal * f( z_lens )

    # Number of galaxies in each radial bin
    Areas = np.pi * (( R_bins[1:] / dl)**2 - ( R_bins[:-1] / dl)**2) * (180 / np.pi * 60)**2
    N_gal = Areas * n_gal

    # Covariance matrix
    sigma_cr = lens.sigma_cr( z_lens, z_lens + 0.5 )  # Critical surface density

    sigma2 = ( 0.3 * sigma_cr )**2 / N_gal

    # add errors to the esd profile
    return -0.5 * np.sum( ( esd_t_obs - esd_t ) ** 2 / sigma2 )
```

```python
def log_prior( param ):
    log10M_lens = param
    if 12 < log10M_lens < 15.5:
        return 0.0
    return -np.inf
```

```python
def log_probability( param , esd_t_obs , z_lens ):
    lp = log_prior( param )
    if not np.isfinite(lp):
        return -np.inf
    return lp + log_likelihood( param , esd_t_obs , z_lens )
```

```python
class Ml_obs_simulator:
    
    def __init__( self , posterior_calculator , ESD_full_profile ):
        """
        Initialize the Universe_simulation class.
        """
        self.posterior_calculator = posterior_calculator
        self.ESD_full_profile = ESD_full_profile
        self.nsamples = 10000
        self.log10m_bins = np.arange( 12 , 15.5 , 0.05 )
        self.log10m_cents = ( self.log10m_bins[1:] + self.log10m_bins[:-1] )/2.

    def run_simulation(self, param_values):
        """
        Run the simulation using the variable parameters provided.
        """
        
        log10M_lens, z_lens = param_values
        z_lens = z_lens.numpy()
        
        # draw shear profile
        esd_t = self.ESD_full_profile.run_simulation( torch.tensor( [ log10M_lens , z_lens ]) )
        
        # estimate mass and variance
        posterior_samples = self.posterior_calculator.sample( ( self.nsamples,), x = esd_t , show_progress_bars = False )
        
        mass_posterior, _ = np.histogram( posterior_samples , bins = log10m_bins )

        return [ np.median( posterior_samples.numpy().T[0] ) , 
                 self.log10m_cents[ np.argmax( mass_posterior )] ,  
                 np.std( posterior_samples.numpy().T[0] ) ] 
```

```python
import torch

def run_many_simulations(input_params, n_simulations):
    """
    Runs multiple simulations for each input parameter.

    Parameters:
    - input_params: A single torch.Tensor of [M_l, z_lens]
    - n_simulations: Number of simulations to run for each input.

    Returns:
    A list of results for each simulation.
    """
    results = []
    for _ in range(n_simulations):
        result = ml_obs_sim.run_simulation(input_params)
        results.append(result)
    
    return results

# Example usage for a single input
n_simulations = 1000  # Number of times to run the simulation for each input
input_params = torch.Tensor([14.5, 0.5])  # Example input [M_l, z_lens]

# Run many simulations for this input
sim_results = run_many_simulations(input_params, n_simulations)

# sim_results will be a list of results for the same input
```

```python
from concurrent.futures import ProcessPoolExecutor
import torch

def run_many_simulations(input_params, n_simulations):
    """
    Runs multiple simulations for each input parameter using parallel processing.
    
    Parameters:
    - input_params: A single torch.Tensor of [M_l, z_lens]
    - n_simulations: Number of simulations to run for each input.
    
    Returns:
    A list of results for each simulation.
    """
    param_list = [input_params] * n_simulations

    # Parallel simulation runs for a single input using ProcessPoolExecutor
    with ProcessPoolExecutor(max_workers=min(n_simulations, 4)) as executor:
        results = list(executor.map(ml_obs_sim.run_simulation, param_list))

    return results

def run_multiple_inputs(input_param_list, n_simulations):
    """
    Runs multiple simulations for each input in parallel using multiple workers.
    
    Parameters:
    - input_param_list: List of input parameters (each a torch.Tensor of [M_l, z_lens])
    - n_simulations: Number of simulations to run for each input.

    Returns:
    A list of results for each input parameter set.
    """
    # Use ProcessPoolExecutor to run simulations for multiple inputs in parallel
    with ProcessPoolExecutor() as executor:
        results = list(executor.map(lambda x: run_many_simulations(x, n_simulations), input_param_list))

    return results

# Example usage for multiple inputs
input_param_list = [
    torch.Tensor([14.5, 0.5]),  # First input
    torch.Tensor([13.8, 0.3]),  # Second input
    torch.Tensor([14.2, 0.4]),  # Third input
    # Add more inputs as needed
]
n_simulations = 100  # Number of simulations per input

# Run simulations for multiple inputs
sim_results_for_multiple_inputs = run_multiple_inputs(input_param_list, n_simulations)

# sim_results_for_multiple_inputs will be a list of lists, 
# where each list contains the results of multiple simulations for a single input
```
