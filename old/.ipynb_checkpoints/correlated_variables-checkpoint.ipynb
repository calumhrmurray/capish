\begin{equation}
S_8 = \sigma_8 \sqrt{\Omega_m/0.3}
\end{equation}

\begin{equation}
\sigma^2(S_8) = (\sqrt{\Omega_m/0.3})^2 \sigma^2(\sigma_8) + (\frac{1}{2}\sqrt{\frac{1}{0.3\Omega_m}}\sigma_8)^2 \sigma^2(\Omega_m) +
\rho (\sqrt{\Omega_m/0.3}) \sigma(\sigma_8) (\frac{1}{2}\sqrt{\frac{1}{0.3\Omega_m}}\sigma_8) \sigma(\Omega_m)
\end{equation}

```python
import numpy as np
```

```python
sigma_om = 2.2*1e-3
sigma_s8 = 1.4 * 1e-3
rho = -0.9
om = 0.3
s8 = 0.8
```

```python
np.sqrt( sigma_om**2 + sigma_s8**2 + rho * sigma_s8 * sigma_om )
```

```python
np.sqrt( sigma_om**2 + sigma_s8**2 + rho * sigma_s8 * sigma_om )
```

```python
corr = np.sqrt( sigma_om**2 + sigma_s8**2 + rho * sigma_s8 * sigma_om )
false = np.sqrt( sigma_om**2 + sigma_s8**2 + 0.95 * rho * sigma_s8 * sigma_om )
(false-corr)/corr
```

```python
corr = np.sqrt( ( 0.5 * s8 / (om * 0.3)**0.5 )**2 * sigma_om**2 + \
                ( om / 0.3) * sigma_s8**2 + \
                rho * sigma_s8 * sigma_om * ( om / 0.3)**0.5 * ( 0.5 * s8 / (om * 0.3)**0.5 ) )

false = np.sqrt( ( 0.5 * s8 / (om * 0.3)**0.5 )**2 * sigma_om**2 + \
                ( om / 0.3) * sigma_s8**2 + \
                0.9 * rho * sigma_s8 * sigma_om * ( om / 0.3)**0.5 * ( 0.5 * s8 / (om * 0.3)**0.5 ) )

(false-corr)/corr
```
